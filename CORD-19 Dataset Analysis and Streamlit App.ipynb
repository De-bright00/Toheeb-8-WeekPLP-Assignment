```python
import pandas as pd
```

```python
df = pd.read_csv('metadata.csv', low_memory=False)
```

```python
df.head()
```

```python
# To Check the number of rows and columns
print("Shape of dataset:", df.shape)
```

```python
## To Display information about each column
print(df.info())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# To Display summary statistics for numerical columns
df.describe()
```

```python
# Check missing data percentage
missing = df.isnull().mean() * 100
print(missing.sort_values(ascending=False))
```

```python
# Drop rows with missing publish_time (critical for time-based analysis)
df = df.dropna(subset=['publish_time'])
```

```python
# Fill missing journals with 'Unknown'
df['journal'] = df['journal'].fillna('Unknown')
```

```python
# Drop unnecessary columns
df = df.drop(columns=['license', 'pmcid', 'sha'], errors='ignore')
```

```python
# Convert publish_time to datetime
df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce')
```

```python
# Extract publication year
df['year'] = df['publish_time'].dt.year
```

```python
# Create new feature: Abstract word count
df['abstract_word_count'] = df['abstract'].fillna('').apply(lambda x: len(x.split()))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
yearly_counts = df['year'].value_counts().sort_index()

plt.figure(figsize=(8,5))
sns.barplot(x=yearly_counts.index, y=yearly_counts.values)
plt.title('COVID-19 Research Publications by Year')
plt.xlabel('Publication Year')
plt.ylabel('Number of Papers')
plt.show()
```

```python
top_journals = df['journal'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(y=top_journals.index, x=top_journals.values)
plt.title('Top 10 Journals Publishing COVID-19 Papers')
plt.xlabel('Number of Papers')
plt.ylabel('Journal')
plt.show()
```

```python
text = ' '.join(df['title'].dropna().tolist())

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Words in Research Titles')
plt.show()
```

```python
source_counts = df['source_x'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(y=source_counts.index, x=source_counts.values)
plt.title('Top 10 Sources of Publications')
plt.xlabel('Count')
plt.ylabel('Source')
plt.show()
```
